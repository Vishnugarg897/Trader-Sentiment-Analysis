# Trader Performance vs Market Sentiment

Load data and run step by step.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sent = pd.read_csv('data/sentiment.csv')
trades = pd.read_csv('data/trades.csv')
```

```python
print(trades.columns)
```

```python
print(sent.columns)
```

```python
trades_small['time'] = pd.to_datetime(trades_small['Timestamp'], errors='coerce')
trades_small['date'] = trades_small['time'].dt.date

daily_pnl = trades_small.groupby('date')['Closed PnL'].sum().reset_index()
```

```python
sent_small = sent[['timestamp','classification']].copy()
sent_small['date'] = pd.to_datetime(sent_small['timestamp']).dt.date
sent_small = sent_small[['date','classification']]
```

```python
daily_trader_pnl = trades[['Account','Closed PnL','Timestamp']].copy()
daily_trader_pnl['time'] = pd.to_datetime(daily_trader_pnl['Timestamp'])
daily_trader_pnl['date'] = daily_trader_pnl['time'].dt.date

daily_trader_pnl = daily_trader_pnl.groupby(['Account','date'])['Closed PnL'].sum().reset_index()
```

```python
sent_small = sent[['timestamp','classification']].copy()
sent_small['date'] = pd.to_datetime(sent_small['timestamp']).dt.date
sent_small = sent_small[['date','classification']]

df2 = daily_trader_pnl.merge(sent_small, on='date', how='inner')
print(df2.shape)
```

```python
sns.boxplot(data=df2, x='classification', y='Closed PnL')
plt.ylim(-2000000, 2000000)   # zoom
plt.show()
```

```python
# mark win / loss
daily_trader_pnl['win'] = daily_trader_pnl['Closed PnL'] > 0

df_win = daily_trader_pnl.merge(sent_small, on='date', how='inner')

df_win.groupby('classification')['win'].mean()
```

```python
#required columns
tmp = trades[['Account','Size USD','Timestamp']].copy()

tmp['date'] = pd.to_datetime(tmp['Timestamp'], errors='coerce').dt.date

# per trader per day average size
daily_size = tmp.groupby(['Account','date'])['Size USD'].mean().reset_index()
```

```python
sent_small = sent[['timestamp','classification']].copy()
sent_small['date'] = pd.to_datetime(sent_small['timestamp'], errors='coerce').dt.date
sent_small = sent_small[['date','classification']].drop_duplicates()
```

```python
df_size = daily_size.merge(sent_small, on='date', how='inner')
print(df_size.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df_size, x='classification', y='Size USD')
plt.ylim(0, df_size['Size USD'].quantile(0.95))
plt.show()
```
